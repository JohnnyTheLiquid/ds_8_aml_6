```python
import numpy as np
import pandas as pd
import Levenshtein as lev

from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
```

**задание 1**

```python
claim_list = [
    'Шехавцова Анна',
    'Гречихина Наталья',
    'Козлова Алена',
    'Груздева Алина',
    'Кущенко Анна',
    'Чистякова Анастасия'
]
```

```python
# результат расшифровки речи диктора

speech_recognition = [
    'кучменко она',
    'кущенко оксана',
    'груздь алина',
    'рычихина наталья',
    'шиховцева на',
    'чистова анастасия'
]
```

```python
for recognized_name in speech_recognition:
    for real_name in claim_list:
        print('{},{},{}'.format(recognized_name, real_name, lev.distance(recognized_name, real_name)))
```

**Решение задания 1:**

```python

for recognized_name in speech_recognition:
    dic = {}
    for real_name in claim_list:
        dic[lev.distance(recognized_name, real_name)] = [recognized_name, real_name]
    print ('Ближайшие к распознанному "{}" имя из списка: "{}", с расстоянием Левенштейна: {}'.format(
        min(dic.items(), key = lambda x: x[0])[1][0],
        min(dic.items(), key = lambda x: x[0])[1][1],
        min(dic.items(), key = lambda x: x[0])[0] ))
```
